```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("lm.csv", encoding_errors='ignore')
df.head()
```

## Анализ сессий по устройствам


```python
m1 = df['Device Category'].value_counts()
m1.head()
```

```python
plt.figure(figsize=(10, 8))

plt.pie(m1.values, autopct='%1.1f%%')

plt.title('Количество сессий по устройствам')
plt.legend(m1.index);
```
